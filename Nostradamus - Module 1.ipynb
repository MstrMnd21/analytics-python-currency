```python
# Nostradamus One: NTS Framework for all pairs.
import pandas as pd
import numpy as np
```

```python
H1FullUsdChf_pd = pd.read_csv('datasets/2018-06-12-UsdChfH1_raw.csv')
H1FullUsdChf_pd

# Rename Columns to: Time, Open, High, Low, Close, Volume
H1FullUsdChf_pd.columns = ['Date', 'Time', 'Open','High','Low','Close', 'Vol']
# H1FullUsdChf_pd

H1FullUsdChf_pd = H1FullUsdChf_pd.set_index('Date')
# H1FullUsdChf_pd
```

```python
T0000UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd['Time'] == '00:00')]
# T2100UsdChf_pd

T0700UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd['Time'] == '07:00')]
# T0400UsdChf_pd
```

```python
T0700UsdChf_pd.columns = ['T07', 'OO7','H07','L07','C07', 'Volume']
T0700UsdChf_pd

T0000UsdChf_pd.columns = ['T00', 'O00','H00','L00','C00', 'Volume']
T0000UsdChf_pd

# Concatenate
T00x04UsdChf_pd = pd.concat([T0700UsdChf_pd, T0000UsdChf_pd], axis=1)

# Reset Index
# T00x04UsdChf_pd = T00x04UsdChf_pd.reset_index()

TableSummary01 = T00x04UsdChf_pd
# TableSummary01
```

```python
D1FullUsdChf_pd = pd.read_csv('datasets/2018-06-12-UsdChfD1_raw.csv')
# dailyFullUsdChfD1_pd = dailyFullUsdChfD1_pd.drop(dailyFullUsdChfD1_pd.columns[[-3, -1]], axis=1)
D1FullUsdChf_pd

# Rename Columns to: Time, Open, High, Low, Close, Volume
D1FullUsdChf_pd.columns = ['Date', 'Time', 'Open','High','Low','Close', 'Vol']
D1FullUsdChf_pd

# Drop columns
D1FullUsdChf_pd = D1FullUsdChf_pd.drop(['Time', 'Close', 'Open', 'Vol'], axis=1)

D1FullUsdChf_pd = D1FullUsdChf_pd.set_index('Date')
D1FullUsdChf_pd

H1xD1_pd = pd.concat([TableSummary01, D1FullUsdChf_pd], axis=1)
H1xD1_pd

# Remove NaN
H1xD1_pd = H1xD1_pd.dropna()
SummaryTable02 = H1xD1_pd
SummaryTable02
```

```python
# Calc0007= SummaryTable02['C07'] - SummaryTable02['C00']

# Change to Open 00 to Close 07
Calc0007= SummaryTable02['C07'] - SummaryTable02['O00']
Calc0007

Calc0007 = pd.DataFrame (Calc0007, columns = ['Calc'])

Calc0007 = Calc0007.reset_index()
Calc0007

BullOrBear = np.where (Calc0007 ['Calc'] > 0, 'bull', 'bear')

BullOrBear = pd.DataFrame (BullOrBear, columns = ['Bull / Bear'])
BullOrBear 

SummaryTable03 = pd.concat([Calc0007, BullOrBear], axis=1)

SummaryTable03.columns = ['Date', 'Calc','Bull/Bear']
SummaryTable03 = SummaryTable03.set_index('Date')

# SummaryTable03 = pd.concat([SummaryTable03, D1FullUsdChf_pd], axis=1)
# SummaryTable03 = SummaryTable03.dropna()
# SummaryTable03

SummaryTable04 = pd.concat([SummaryTable03, SummaryTable02], axis=1)
SummaryTable04 = SummaryTable04.drop(['H07', 'L07', 'H00', 'L00', 'Volume'], axis=1)
SummaryTable04
```

```python
# If Buy or If Sell 
mDataPositionBuy = SummaryTable04 ['High'] - SummaryTable04 ['C07']
mDataPositionBuy_pd = pd.DataFrame (mDataPositionBuy, columns = ['If Buy Pos'])
# mDataPositionBuy_pd

mDataPositionSell = SummaryTable04 ['C07'] - SummaryTable04 ['Low']
mDataPositionSell_pd = pd.DataFrame (mDataPositionSell, columns = ['If Sell Pos'])

# Print 
# mDataPositionSell_pd 
# mDataPositionBuy_pd

IfBuyIfSell_pd = pd.concat([mDataPositionBuy_pd, mDataPositionSell_pd], axis=1)
IfBuyIfSell_pd

TableSummary05 = pd.concat([IfBuyIfSell_pd, SummaryTable04], axis=1)
TableSummary05

TableSummary05 = TableSummary05.drop(['Calc', 'T07','High', 'Low'], axis=1)

TableSummary05 = TableSummary05.drop(['C00', 'OO7', 'C07', 'T00', 'O00'], axis=1)
TableSummary05
# Drop NaN
# IfBuyIfSell_pd = IfBuyIfSell_pd.dropna()
# IfBuyIfSell_pd

# summaryTable02 = pd.concat([dailyFullUsdChfD1_pd, IfBuyIfSell_pd, BullOrBearResult_pd], axis=1)
# summaryTable02
```

```python
# ** Capture all bull and capture all bear

tableForBuyEntry_pd = TableSummary05[(TableSummary05['Bull/Bear'] == 'bull')]
tableForSellEntry_pd = TableSummary05[(TableSummary05['Bull/Bear'] == 'bear')]

# Print
# tableForBuyEntry_pd
# tableForSellEntry_pd
```

```python
# For buy position result
BuyPosSubResult = np.where (tableForBuyEntry_pd ['If Buy Pos'] > 0.0008, 'Success', 'Fail')
BuyPosSubResult_pd = pd.DataFrame (BuyPosSubResult, columns = ['Buy Result'])
BuyPosSubResult_pd  

# Convert To Numpy 
BuyPosSubResult_np = BuyPosSubResult_pd.values
BuyPosSubResult_pd

# Test count success 
# (BuyPosSubResult_np == 'Success').sum() 

# Count Contents
countTotalResultForBuy = len(BuyPosSubResult_np)
countSuccessResultBuy = (BuyPosSubResult_np == 'Success').sum() 

WinRateForBuyPos = countSuccessResultBuy / countTotalResultForBuy
WinRateForBuyPos
```

```python
SellPosSubResult = np.where (tableForSellEntry_pd ['If Sell Pos'] > 0.0008, 'Success', 'Fail')
SellPosSubResult_pd = pd.DataFrame (SellPosSubResult, columns = ['Sell Result'])
SellPosSubResult_pd  

# Convert To Numpy
SellPosSubResult_np = SellPosSubResult_pd.values
(SellPosSubResult_np == 'Success').sum()

# Count Contents
countResultForSell = len(SellPosSubResult_np)
countSuccessResultSell = (SellPosSubResult_np == 'Success').sum() 

WinRateForSell = countSuccessResultSell / countResultForSell
WinRateForSell
```
