```python
# Attach all dataset and convert to dataframe
hourlyFullUsdChf_pd = pd.read_csv('UsdChfH1NoonTime.csv')
dailyFullUsdChf_pd = pd.read_csv('UsdChfD1NoonTime.csv')
hourlySniffHighUsdChf_pd = pd.read_csv('UsdChfH1NoonTimeForSniffingHigh.csv')

# Noon Time Range is 21:00 to 04:00 (6am to 1pm Philippine Time). Create a dataframe for 21:00 and 04:00
mDataTime2100_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '21:00')]
mDataTime0400_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '04:00')] 

# Delete Columns
mDataTime0400_pd = mDataTime0400_pd.drop(mDataTime0400_pd.columns[[5]], axis=1)
mDataTime2100_pd = mDataTime2100_pd.drop(mDataTime2100_pd.columns[[5]], axis=1)

# Daily High Low
mDataHighDaily_pd = dailyFullUsdChf_pd.drop(dailyFullUsdChf_pd.columns[[0, 1, 4, 5]], axis=1)

# Change Column Names
mDataTime0400_pd.columns = ['Time 04', 'Open 04','High 04','Low 04','Close 04']
mDataTime2100_pd.columns = ['Time 21', 'Open 21','High 21','Low 21','Close 21']
mDataHighDaily_pd.columns = ['DHigh', 'DLow']

# Concatenate dataframes at vertical axis 
# Reshaped Dataset
mDataTime21To04ToDHigh_pd = pd.concat([mDataTime2100_pd, mDataTime0400_pd, mDataHighDaily_pd], axis=1)
reshapedDataSet_pd = mDataTime21To04ToDHigh_pd

# Notes: After reshaping the dataset, we will dataframe analytics
```
