```python
# PREMONITION by ANALYTICS
# Create table for entry direction decisions
# Price change from Open 00 to Close 03

EntryDirection = Table_Horizontal_date['C03'] - Table_Horizontal_date['O00']
EntryDirection = pd.DataFrame (EntryDirection, columns = ['CALC'])
EntryDirection['POS'] = 'Default'
EntryDirection['POS'][EntryDirection['CALC'] > 0] = 'Long'
EntryDirection['POS'][EntryDirection['CALC'] < 0] = 'Short'

EntryDirection_date = EntryDirection
```

```python
# Calculate HIghs and Lows for 04:00 to 10:00

High_04TO10 = Table_Horizontal_date [['H04','H05','H06','H07', 'H08', 'H09', 'H10']]
High_04TO10['H0410'] = High_04TO10.max(axis=1)

Low_04TO10 = Table_Horizontal_date [['L04','L05','L06','L07', 'L08', 'L09', 'L10']]
Low_04TO10['L0410'] = Low_04TO10.min(axis=1)
```

```python
# Create table for Closing 03:00
UsdJpyC03 = T03UsdJpy_date[['C03']]
UsdJpyO00 = T00UsdJpy_date[['O00']]
```

```python
# Concatenate the tables
Table_HighLowClosePosition = pd.concat([UsdJpyO00, High_04TO10, Low_04TO10, UsdJpyC03, EntryDirection_date], axis=1)
Table_HighLowClosePosition
# Cleanup
# Table_HighLowClosePosition = Table_HighLowClosePosition[['H0410', 'L0410','C03', 'POS']]
# Table_HighLowClosePosition
```

```python
# Calculate Success Rate

LongPosition = Table_HighLowClosePosition ['H0410'] - Table_HighLowClosePosition ['C03']
LongPosition = pd.DataFrame (LongPosition, columns = ['PIP LONG'])

ShortPosition = Table_HighLowClosePosition ['C03'] - Table_HighLowClosePosition ['L0410']
ShortPosition = pd.DataFrame (ShortPosition, columns = ['PIP SHORT'])
```

```python
Table_HighLowClosePosition = pd.concat([Table_HighLowClosePosition, LongPosition, ShortPosition], axis=1)
Table_HighLowClosePosition
```

```python
# Capture all Long and Short Positions

SummaryLongPosition = Table_HighLowClosePosition[(Table_HighLowClosePosition['POS'] == 'Long')]
SummaryShortPosition = Table_HighLowClosePosition[(Table_HighLowClosePosition['POS'] == 'Short')]
```

```python
# For long position result
LongPosSubResult = np.where (SummaryLongPosition ['PIP LONG'] > 0.009, 'Success', 'Fail')
LongPosSubResult = pd.DataFrame (LongPosSubResult, columns = ['LONG RESULT'])
LongPosSubResult

# Isolate BuyPosSubResult_pd
LongPosSubResult_x = LongPosSubResult

# Convert To Numpy 
LongPosSubResult_np = LongPosSubResult_x.values

# Count Contents
countTotalResultForLong = len(LongPosSubResult_np)
countSuccessResultLong = (LongPosSubResult_np == 'Success').sum() 

WinRateForLongPos = countSuccessResultLong / countTotalResultForLong
WinRateForLongPos
```

```python
ShortPosSubResult = np.where (SummaryShortPosition ['PIP SHORT'] > 0.009, 'Success', 'Fail')
ShortPosSubResult = pd.DataFrame (ShortPosSubResult, columns = ['SHORT RESULT'])
ShortPosSubResult  

# Isolate SellPosSubResult_pd
ShortPosSubResult_x = ShortPosSubResult

# Convert To Numpy
ShortPosSubResult_np = ShortPosSubResult_x.values

# Count Contents
countTotalResultForShort = len(ShortPosSubResult_np)
countSuccessResultShort = (ShortPosSubResult_np == 'Success').sum() 

WinRateForShortPos = countSuccessResultShort / countTotalResultForShort
WinRateForShortPos
```
