```python
# Nostradamus One: NTS Framework for all pairs.
import pandas as pd
import numpy as np
```

```python
hourlyFullUsdChf_pd = pd.read_csv('datasets/2018-05-02-UsdChfH1.csv')
```

```python
mDataTime2100_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '21:00')]
```

```python
mDataTime0400_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '04:00')]
mDataTime0400_pd.columns = ['T04', 'OO4','H04','L04','C04', 'V04']


mDataTime2100_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '21:00')]
mDataTime2100_pd.columns = ['T21', 'O21','H21','L21','C21', 'V21']

```

```python
concat0421Table = pd.concat([mDataTime0400_pd, mDataTime2100_pd], axis=1)
concat0421Table
```

```python
mDataBullBear = concat0421Table ['C21'] - concat0421Table ['C04']
# mDataBullBear

mDataBullBear_pd = pd.DataFrame (mDataBullBear, columns = ['BlBr Calc'])
# mDataBullBear_pd

concat0421BlBrTable_pd = pd.concat([concat0421Table, mDataBullBear_pd], axis=1)
# concat0421BlBrTable_pd

mDataBullBearResult = np.where (concat0421BlBrTable_pd ['BlBr Calc'] > 0, 'bull', 'bear')
mDataBullBearResult_pd = pd.DataFrame (mDataBullBearResult, columns = ['Bull or Bear'])

# mDataBullBearResult_pd 
concat0421BlBrResTable_pd = pd.concat([concat0421BlBrTable_pd, mDataBullBearResult_pd], axis=1)
concat0421BlBrResTable_pd
```

```python
mDataBullBearResult_pd
```

```python
dailyFullUsdChfD1_pd = pd.read_csv('datasets/2018-05-02-UsdChfD1.csv')
dailyFullUsdChfD1_pd
dailyFullUsdChfD1_pd.columns = ['0', '0','DHigh','DLow','0', '0']
dailyFullUsdChfD1_pd
```

```python
# If Buy of If Sell 
mDataPositionBuy = dailyFullUsdChfD1_pd ['DHigh'] - concat0421Table ['C04']
mDataPositionBuy_pd = pd.DataFrame (mDataPositionBuy, columns = ['If Buy'])
# Print
# DataPositionBuy_pd

mDataPositionSell = concat0421Table ['C04'] - dailyFullUsdChfD1_pd ['DLow']
mDataPositionSell_pd = pd.DataFrame (mDataPositionSell, columns = ['If Sell'])
# Print 
# mDataPositionSell_pd 

concatIfBuySell_pd = pd.concat([mDataPositionBuy_pd, mDataPositionSell_pd], axis=1)
# Print
# concatIfBuySell_pd

# Convert to Numpy
concatIfBuySell_np = concatIfBuySell_pd.values
# concatIfBuySell_pd

mDataBullBearResult_np = mDataBullBearResult_pd.values
# Print
# mDataBullBearResult_pd

concatIfBuySell_pd = concatIfBuySell_pd.reset_index(drop=True)
concatIfBuySell_pd

concatIfBuySellBlBrRes_pd = pd.concat([concatIfBuySell_pd, mDataBullBearResult_pd], axis=1)
# Print
# concatIfBuySellBlBrRes_pd

# Assign rename to more readable variable
summaryIfBuySellBullBearResult_pd = concatIfBuySellBlBrRes_pd.dropna()

# Print
summaryIfBuySellBullBearResult_pd
```

```python
# ** Capture all bull and capture all bear

# Bull
summaryBuyEntry_pd = summaryIfBuySellBullBearResult_pd[(summaryIfBuySellBullBearResult_pd['Bull or Bear'] == 'bull')]

# Print
# summaryBuyEntry_pd

# Bear
summarySellEntry_pd = summaryIfBuySellBullBearResult_pd[(summaryIfBuySellBullBearResult_pd['Bull or Bear'] == 'bear')]

# Print
```

```python
BuyPosSubResult = np.where (summaryBuyEntry_pd ['If Buy'] > 0.000500, 'Success', 'Fail')
BuyPosSubResult_pd = pd.DataFrame (BuyPosSubResult, columns = ['Buy Result'])
BuyPosSubResult_pd  

# concatIfBuySellBlBrRes_pd
```

```python
# Convert To Numpy 
BuyPosSubResult_np = BuyPosSubResult_pd.values

# Test count success 
(BuyPosSubResult_np == 'Success').sum() 

# Count Contents
countResultForBuy = len(BuyPosSubResult_np)
countSuccessResultBuy = (BuyPosSubResult_np == 'Success').sum() 

WinRateForBuy = countSuccessResultBuy / countResultForBuy
WinRateForBuy
# countSuccessResultBuy
```

```python
SellPosSubResult = np.where (summarySellEntry_pd ['If Sell'] > 0.000500, 'Success', 'Fail')
SellPosSubResult_pd = pd.DataFrame (SellPosSubResult, columns = ['Sell Result'])
SellPosSubResult_pd  

# Convert To Numpy
SellPosSubResult_np = SellPosSubResult_pd.values
(SellPosSubResult_np == 'Success').sum()

# Count Contents
countResultForSell = len(SellPosSubResult_np)
countSuccessResultSell = (SellPosSubResult_np == 'Success').sum() 

WinRateForSell = countSuccessResultSell / countResultForSell
WinRateForSell

# Note: Needs to drop the NaN within the column. 
```
