```python
# Nostradamus One: NTS Framework for all pairs.
import pandas as pd
import numpy as np
```

```python
hourlyFullUsdChf_pd = pd.read_csv('datasets/2018-05-02-UsdChfH1.csv')
```

```python
mDataTime2100_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '21:00')]
```

```python
mDataTime0400_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '04:00')]
mDataTime0400_pd.columns = ['T04', 'OO4','H04','L04','C04', 'V04']


mDataTime2100_pd = hourlyFullUsdChf_pd [(hourlyFullUsdChf_pd['Time'] == '21:00')]
mDataTime2100_pd.columns = ['T21', 'O21','H21','L21','C21', 'V21']

```

```python
concat0421Table = pd.concat([mDataTime0400_pd, mDataTime2100_pd], axis=1)
concat0421Table
```

```python
mDataBullBear = concat0421Table ['C21'] - concat0421Table ['C04']
# mDataBullBear

mDataBullBear_pd = pd.DataFrame (mDataBullBear, columns = ['BlBr Calc'])
# mDataBullBear_pd

concat0421BlBrTable_pd = pd.concat([concat0421Table, mDataBullBear_pd], axis=1)
# concat0421BlBrTable_pd

mDataBullBearResult = np.where (concat0421BlBrTable_pd ['BlBr Calc'] > 0, 'bull', 'bear')
mDataBullBearResult_pd = pd.DataFrame (mDataBullBearResult, columns = ['Bull or Bear'])

# mDataBullBearResult_pd 
concat0421BlBrResTable_pd = pd.concat([concat0421BlBrTable_pd, mDataBullBearResult_pd], axis=1)
concat0421BlBrResTable_pd
```
