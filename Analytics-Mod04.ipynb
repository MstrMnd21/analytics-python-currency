```python
# Selecting Columns for max value between H08 TO H14
tableH08to14 = SummaryTable06[['H08','H09','H10','H11','H12', 'H13', 'H14']]
tableH08to14['High 08to14'] = tableH08to14.max(axis=1)
tableH08to14
```

```python
tableL08to14 = SummaryTable06[['L08','L09','L10','L11','L12', 'L13','L14']]
tableL08to14['Low 08to14'] = tableL018to14.min(axis=1)
tableL08to14
```

```python
# Creating columns for Close 07
tableC07 = SummaryTable06[['C07']]
tableC07
```

```python
# Entry Direction Calc
EntryCalc = SummaryTable06['C07'] - SummaryTable06['O00']
EntryCalc = pd.DataFrame (EntryCalc, columns = ['Entry Calc'])
EntryCalc = EntryCalc.reset_index()

BullOrBear = np.where (EntryCalc ['Entry Calc'] > 0, 'bull', 'bear')
BullOrBear = pd.DataFrame (BullOrBear, columns = ['Bull / Bear'])

SummaryTable03 = pd.concat([EntryCalc, BullOrBear], axis=1)
SummaryTable03.columns = ['Date', 'Entry Calc','Bull/Bear']
SummaryTable03 = SummaryTable03.set_index('Date')

tableBlBr = SummaryTable03[['Bull/Bear']]
```

```python
summaryTable07 = pd.concat([tableC07, tableH08to14, tableL08to14, tableBlBr], axis=1)
summaryTable07
```

```python
# If Buy or If Sell 
PositionBuy_TP0814 = summaryTable07 ['High 08to14'] - summaryTable07 ['C07']
PositionBuy_TP0814_pd = pd.DataFrame (PositionBuy_TP0814, columns = ['If Buy Pos'])

PositionSell_TP0814 = summaryTable07 ['C07'] - summaryTable07 ['Low 08to14']
PositionSell_TP0814_pd = pd.DataFrame (PositionSell_TP0814, columns = ['If Sell Pos'])

summaryTable0814 = pd.concat([PositionBuy_TP0814_pd, PositionSell_TP0814_pd, tableBlBr], axis=1)
summaryTable0814
```

```python
tableForBuyEntry_0814_pd = summaryTable0814[(summaryTable0814['Bull/Bear'] == 'bull')]
tableForSellEntry_0814_pd = summaryTable0814[(summaryTable0814['Bull/Bear'] == 'bear')]
tableForSellEntry_0814_pd
```

```python
# For buy position result
BuyPosSubResult_0814 = np.where (tableForBuyEntry_0814_pd ['If Buy Pos'] > 0.0008, 'Success', 'Fail')
BuyPosSubResult_0814_pd = pd.DataFrame (BuyPosSubResult_0814, columns = ['Buy Result'])
BuyPosSubResult_0814_pd  

# Convert To Numpy 
BuyPosSubResult_0814_np = BuyPosSubResult_0814_pd.values
BuyPosSubResult_0814_pd

# Count Contents
countTotalResultForBuy_0814 = len(BuyPosSubResult_0814_np)
countSuccessResultBuy_0814 = (BuyPosSubResult_0814_np == 'Success').sum() 

WinRateForBuyPos_0814 = countSuccessResultBuy_0814 / countTotalResultForBuy_0814
WinRateForBuyPos_0814
```

```python
# For buy position result
SellPosSubResult_0814 = np.where (tableForBuyEntry_0814_pd ['If Sell Pos'] > 0.00080, 'Success', 'Fail')
SellPosSubResult_0814_pd = pd.DataFrame (SellPosSubResult_0814, columns = ['Sell Result'])
SellPosSubResult_0814_pd  

# Convert To Numpy 
SellPosSubResult_0814_np = SellPosSubResult_0814_pd.values
SellPosSubResult_0814_pd

# Count Contents
countTotalResultForSell_0814 = len(SellPosSubResult_0814_np)
countSuccessResultSell_0814 = (SellPosSubResult_0814_np == 'Success').sum() 

WinRateForSellPos_0814 = countSuccessResultSell_0814 / countTotalResultForSell_0814
WinRateForSellPos_0814
```
