```python
# Nostradamus One: NTS Framework for all pairs.
import pandas as pd
import numpy as np
```

```python
# TP WIN/LOSS RATE 0:00 -> 09:00 (Noxtradamus V2)
# Include HIGH AND LOW FOR 10:00 AND 11:00 | H10 L10 & H11 L11
H1FullUsdChf_pd = pd.read_csv('datasets/2018-06-12-UsdChfH1_raw.csv')

H1FullUsdChf_pd.columns = ['Date', 'Time', 'Open','High','Low','Close', 'Vol']
H1FullUsdChf_pd = H1FullUsdChf_pd.set_index('Date')
H1FullUsdChf_pd = H1FullUsdChf_pd.drop(['Vol'], axis=1)
H1FullUsdChf_pd
```

```python
T0000UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd ['Time'] == '00:00')]

T0900UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd ['Time'] == '09:00')]
T0100UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd ['Time'] == '10:00')]
T0110UsdChf_pd = H1FullUsdChf_pd [(H1FullUsdChf_pd ['Time'] == '11:00')]

T0000UsdChf_pd.columns = ['T00', 'O00','H00','L00','C00']


T0900UsdChf_pd.columns = ['T09', 'O09','H09','L09','C09']
T0100UsdChf_pd.columns = ['T10', 'O10','H10','L10','C10']
T0110UsdChf_pd.columns = ['T11', 'O11','H11','L11','C11']


SummaryTable01 = pd.concat([T0000UsdChf_pd, T0900UsdChf_pd, T0100UsdChf_pd, T0110UsdChf_pd], axis=1)
SummaryTable01
```

```python
# Selecting Columns for max value between H08 and H09
tableH10to11 = SummaryTable01[['H10','H11']]
tableH10to11['High 10To11'] = tableH10to11.max(axis=1)
tableH10to11
```

```python
tableL10to11 = SummaryTable01[['L10','L11']]
tableL10to11['Low 10To11'] = tableL10to11.min(axis=1)
tableL10to11
```

```python
# Creating columns for Close 09
tableC09 = SummaryTable01[['C09']]
tableC09
```

```python
# ENTRY CALCULATION

EntryCalc = SummaryTable01['C09'] - SummaryTable01['C00']
EntryCalc = pd.DataFrame (EntryCalc, columns = ['Entry Calc'])
EntryCalc = EntryCalc.reset_index()

BullOrBear = np.where (EntryCalc ['Entry Calc'] > 0, 'bull', 'bear')
BullOrBear = pd.DataFrame (BullOrBear, columns = ['Bull / Bear'])

SummaryTable02 = pd.concat([EntryCalc, BullOrBear], axis=1)

SummaryTable02.columns = ['Date', 'Entry Calc','Bull/Bear']
SummaryTable02 = SummaryTable02.set_index('Date')

tableBlBr = SummaryTable02[['Bull/Bear']]
tableBlBr
```

```python
summaryTable03 = pd.concat([tableC09, tableH10to11, tableL10to11, tableBlBr], axis=1)
summaryTable03
```

```python
# If Buy or If Sell 
PositionBuy_TP1011 = summaryTable03 ['High 10To11'] - summaryTable03 ['C09']
PositionBuy_TP1011_pd = pd.DataFrame (PositionBuy_TP1011, columns = ['If Buy Pos'])

PositionSell_TP1011 = summaryTable03 ['C09'] - summaryTable03 ['Low 10To11']
PositionSell_TP1011_pd = pd.DataFrame (PositionSell_TP1011, columns = ['If Sell Pos'])

summaryTable04 = pd.concat([PositionBuy_TP1011_pd, PositionSell_TP1011_pd, tableBlBr], axis=1)
summaryTable04
```

```python
tableForBuyEntry_1011_pd = summaryTable04[(summaryTable04['Bull/Bear'] == 'bull')]
tableForSellEntry_1011_pd = summaryTable04[(summaryTable04['Bull/Bear'] == 'bear')]
tableForSellEntry_1011_pd
```

```python
# For buy position result
BuyPosSubResult_1011 = np.where (tableForBuyEntry_1011_pd ['If Buy Pos'] > 0.0006, 'Success', 'Fail')
BuyPosSubResult_1011_pd = pd.DataFrame (BuyPosSubResult_1011, columns = ['Buy Result'])
BuyPosSubResult_1011_pd  

# Convert To Numpy 
BuyPosSubResult_1011_np = BuyPosSubResult_1011_pd.values
BuyPosSubResult_1011_pd

# Count Contents
countTotalResultForBuy_1011 = len(BuyPosSubResult_1011_np)
countSuccessResultBuy_1011 = (BuyPosSubResult_1011_np == 'Success').sum() 

WinRateForBuyPos_1011 = countSuccessResultBuy_1011 / countTotalResultForBuy_1011
WinRateForBuyPos_1011
```

```python
# FOR SELL POSITION RESULT
SellPosSubResult_1011 = np.where (tableForBuyEntry_1011_pd ['If Sell Pos'] > 0.00060, 'Success', 'Fail')
SellPosSubResult_1011_pd = pd.DataFrame (SellPosSubResult_1011, columns = ['Sell Result'])
SellPosSubResult_1011_pd  

# Convert To Numpy 
SellPosSubResult_1011_np = SellPosSubResult_1011_pd.values
SellPosSubResult_1011_pd

# Count Contents
countTotalResultForSell_1011 = len(SellPosSubResult_1011_np)
countSuccessResultSell_1011 = (SellPosSubResult_1011_np == 'Success').sum() 

WinRateForSellPos_1011 = countSuccessResultSell_1011 / countTotalResultForSell_1011
WinRateForSellPos_1011
```
