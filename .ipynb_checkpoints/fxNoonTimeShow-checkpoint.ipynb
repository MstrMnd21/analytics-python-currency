```python
# Nostradamus One: Noon Time Show System
import pandas as pd
import numpy as np
```

```python
mDataFullUsdChfH1_pd = pd.read_csv('USDCHFH1.csv')
mDataFullUsdChfD1_pd = pd.read_csv('UsdChfD1Clean101.csv')
```

```python
mDataFullUsdChfH1_pd
```

```python
# Noon Time Range is 21:00 to 04:00 (6am to 1pm Philippine Time)
mDataTime2100_pd = mDataFullUsdChfH1_pd [(mDataFullUsdChfH1_pd['Time'] == '21:00')]
mDataTime2100_pd

```

```python
mDataTime0400_pd = mDataFullUsdChfH1_pd [(mDataFullUsdChfH1_pd['Time'] == '04:00')] 
```

```python
mDataTime0400_pd
```

```python
#Change Column Names
mDataTime0400_pd.columns = ['Time 04', 'Open 04','High 04','Low 04','Close 04','Volume 04']
mDataTime2100_pd.columns = ['Time 21', 'Open 21','High 21','Low 21','Close 21','Volume 21']
```

```python
# Concatenate dataframes at vertical axis
result = pd.concat([mDataTime2100_pd, mDataTime0400_pd], axis=1)
result
```
